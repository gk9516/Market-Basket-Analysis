```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from csv import reader
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
groceries = []
with open ('/content/drive/MyDrive/basket_analysis.csv','r') as read_obj:
  csv_reader = reader(read_obj)
  for row in csv_reader:
    groceries.append(row)
groceries[0:5]
```

```python
encoder= TransactionEncoder()
transactions = encoder.fit(groceries).transform(groceries)
transactions
```

```python
transactions = transactions.astype('int')
transactions
```

```python
df = pd.DataFrame(transactions, columns=encoder.columns_)
```

```python
df.head()
```

```python
df.shape
```

```python
frequent_itemsets = apriori(df, min_support=0.02, use_colnames= True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x:len(x))
frequent_itemsets
```

```python
frequent_itemsets[(frequent_itemsets['length'] > 1) & (frequent_itemsets['support'] >= 0.05)]
```

```python
frequent_itemsets[(frequent_itemsets['length'] == 2) & (frequent_itemsets['support'] >= 0.02)]
```

```python
rules = association_rules(frequent_itemsets,metric = 'support', min_threshold = 0.02)
rules
```

```python
rules.sort_values(by='confidence',ascending = False)[0:10]
```

```python
rules[(rules['support'] >= 0.02 ) & (rules['lift'] > 1.0)]
```
